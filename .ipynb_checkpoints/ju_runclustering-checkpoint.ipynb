```python
#%run utility_getrechits.py
%matplotlib inline
from HGCalImageAlgo3D import *
plt.style.use('classic')
plt.rc("figure",facecolor="w",figsize=(6,4))
plt.rc("font",size=10)
plt.rc("savefig",dpi=300)
```

```python
#DatasetFile   = 'CMSSW9304_partGun_PDGid22_x100_E30.0To30.0_NTUP'
DatasetFile   = 'CMSSW9304_partGun_PDGid22_x1000_E300.0To300.0_NTUP'
df = pd.read_pickle("../data/input/"+DatasetFile+"_rechit.pkl") 
```

```python
dfresultrech  = pd.DataFrame()
dfresultclus  = pd.DataFrame()

for ievt in np.unique(df.id):
    if ievt%100== 0:
        print(ievt)
    dfevtrech,dfevtclus = ImageAlgorithm(df,ievt)
    dfresultrech = dfresultrech.append(dfevtrech, ignore_index=True)
    dfresultclus = dfresultclus.append(dfevtclus, ignore_index=True)
dfresultrech.to_pickle("../data/output/"+DatasetFile+"_OutputRech.pkl")
dfresultclus.to_pickle("../data/output/"+DatasetFile+"_OutputClus.pkl")
```
